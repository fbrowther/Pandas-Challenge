```python
#load the dependencies
import pandas as pd
import numpy as np
```

```python
#data files to analyse
schools_data_to_load = "/Users/Farzana/Bootcamp UB/Pandas-Challenge/Resources/schools_complete.csv"
students_data_to_load = "/Users/Farzana/Bootcamp UB/Pandas-Challenge/Resources/students_complete.csv"
```

```python
#read_csv
df_schools = pd.read_csv(schools_data_to_load)
df_students = pd.read_csv(students_data_to_load)
```

```python
df_schools.head()
```

```python
df_students.head()
```

```python
#Combine the data into a single file  
df_complete = pd.merge(df_students, df_schools, how="left", on=["school_name", "school_name"])
```

```python
df_complete.head()
```

```python
# Calculate the number of schools in the DateFrame 
Number_of_schools = len(df_complete["school_name"].unique())

# Calculate the number of students in the DataFrame
Total_number_of_students = df_complete["student_name"].count()

print(Number_of_schools)
print(Total_number_of_students)
```

```python
# Calculate the total budget of the schools
Total_budget = df_schools["budget"].sum()
print(Total_budget)
```

```python
# Calculate the average math scores of the students
df_complete["math_score"].mean()
```

```python
# Calculate the average reading scores of the students
df_complete["reading_score"].mean()
```

```python
# The no. of students who passed reading
df_reading_pass = df_complete.loc[df_complete["reading_score"]>= 70,:]
df_reading_pass
```

```python
# The percentage of students who passed reading
Percent_reading_pass = len(df_reading_pass)/len(df_complete)*100
print(Percent_reading_pass)
```

```python
# The no. of students who passed math
df_math_pass = df_complete.loc[df_complete["math_score"]>= 70,:]
df_math_pass.head()
```

```python
# The percentage of students who passed math
Percent_math_pass = len(df_math_pass)/len(df_complete)*100
print(Percent_math_pass)
```

```python
# The no. of students who passed both 
df_pass_both = df_complete.loc[(df_complete["reading_score"]>=70) & (df_complete["math_score"]>=70)]
```

```python
# The percentage of students who passed both
Percent_pass_both = len(df_pass_both)/len(df_complete)*100
print(Percent_pass_both)
```

```python
# School performances based on average spending ranges per student  
bins=[0, 598, 624, 648, 675]
labels=['<$599', '$600-625', '$626-649', '$650-675']
```

```python
# Categorize spending based on the bin.
School_summary["Spending ranges"]= pd.cut(Budget_Per_Student, bins, labels=labels)
```
